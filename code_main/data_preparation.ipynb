## Chose stock code and column atribute
The stock code chosen is the company in variant sector who have high rank. The list of the company is:
1. BBCA (from Financials Sector)
2. AMMN (from Basic Materials Sector)
3. TLKM (from Infrastructures Sector)
4. ASII (from Industrials Sector)
5. AMRT (from Consumer Non-Cyclicals Sector)
6. GOTO (from Technology Sector)
7. ADRO (from Energy Sector)
8. KLBF (from Healthcare Sector)
9. MAPI (from Consumer Cyclicals Sector)


Based on the example of stock price EDA at 
https://www.kaggle.com/code/faressayah/stock-market-analysis-prediction-using-lstm#6.-Predicting-the-closing-price-stock-price-of-APPLE-inc
The attribute needed is compare with raw data column:
- Date :  Tanggal Perdagangan Terakhir
- Stock Code : Kode Saham
- Open : Sebelumnya (Because using Open Price have some null value, 'Sebelumnya' is more complete value)
- High : Tertinggi
- Low : Terendah
- Close : Penutupan
- Adj Close: -
- Volume : Volume

Adjusted close (Adj Close) column doesn't have the equivalent column. So the column taken is just for Date, Stock Code, Open, High, Low, Close, and Volume.

In conclusion the stock code taken is:
[BBCA, AMMN, TLKM, ASII, AMRT, GOTO, ADRO, KLBF, MAPI]

The attribute taken at raw data is:
[Tanggal, Kode Saham, Open Price, Tertinggi, Terendah, Penutupan, Volume]

The attribute will be renamed with english term with code friendly format:
[date, stock_code, open, high, low, close, volume]


```python
#chose stock code
chosen = ['BBCA', 'AMMN', 'TLKM', 'ASII', 'AMRT', 'GOTO', 'ADRO', 'KLBF', 'MAPI']
list_pred = []
i = 1

#iterate in year list
for file_year in [file2024, file2023, file2022, file2021, file2020]:
  print("Year", i)
  j = 1
  i=i+1
  #see every day progress
  for day_stock in file_year:
    print("Day - ", j)
    j = j + 1
    sampel = pd.read_excel(path_raw_data+'/'+day_stock)
    chosen_df = sampel[sampel['Kode Saham'].isin(chosen)]
    after_chose_df = chosen_df[['Tanggal Perdagangan Terakhir','Kode Saham', 'Sebelumnya', 'Tertinggi', 'Terendah', 'Penutupan', 'Volume']]
    list_pred.append(after_chose_df)

```

## Merge all the data every files

```python
result_df = pd.concat(list_pred)
```

## See first 9 data

```python
result_df.head(9)
```

## Rename column

```python
result_df = result_df.rename(columns={'Tanggal Perdagangan Terakhir' : 'date', 'Kode Saham' : 'stock_code',
                          'Sebelumnya' : 'open',  'Tertinggi' : 'high', 'Terendah' : 'low', 'Penutupan' : 'close', 'Volume' : 'volume'})
```

```python
result_df.head(9)
```

## Convert all date format to yyyy-mm-dd

```python
def conv_month_to_num(str):
  if str == 'Jan':
    return 1
  elif str == 'Feb':
    return 2
  elif str == 'Mar':
    return 3
  elif str == 'Apr':
    return 4
  elif str == 'Mei':
    return 5
  elif str == 'Jun':
    return 6
  elif str == 'Jul':
    return 7
  elif str == 'Agt':
    return 8
  elif str == 'Sep':
    return 9
  elif str == 'Okt':
    return 10
  elif str == 'Nov':
    return 11
  elif str == 'Des':
    return 12

```

```python
len_list = len(result_df)
for num in range(len_list):
  value_date = result_df['date'].iloc[num]
  #print(value_date)
  list_date = value_date.split()
  date = int(list_date[0])
  month = conv_month_to_num(list_date[1])
  year = int(list_date[2])
  #print(date, month, year)
  obj_date = datetime.datetime(year, month, date)
  result_df['date'].iloc[num] = obj_date.date()
```

## Change date data as index of the table and sort ascending

```python
result_df.set_index('date', inplace=True)
result_df.sort_values(by='date', ascending = True, inplace = True)
```

## Split data based on stock code and save at the direktory

```python
df_stock = result_df
for stock in chosen:
  df_stock = result_df[result_df['stock_code'] == stock]
  first_column = df_stock.pop('stock_code')
  df_stock.insert(0, 'stock_code', first_column)
  #menyimpan file setiap emiten ke dalam format csv
  df_stock.to_csv('../data_after_prep/'+stock+'2001-2407.csv')
```
